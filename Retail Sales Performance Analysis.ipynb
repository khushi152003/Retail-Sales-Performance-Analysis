# Pareto Analysis (80/20 Rule)

```python
customer_sales = (
    df.groupby('CustomerID')['TotalSales']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

customer_sales.head()
```

```python
customer_sales['Revenue %'] = (
    customer_sales['TotalSales']
    / customer_sales['TotalSales'].sum()
) * 100
```

```python
customer_sales['Cumulative Revenue %'] = (
    customer_sales['Revenue %']
    .cumsum()
)

customer_sales.head()
```

```python
plt.figure(figsize=(14,6))

plt.plot(
    customer_sales['Cumulative Revenue %'],
    color='red',
    linewidth=2
)

plt.axhline(80, linestyle='--')

plt.title("Pareto Analysis (80/20 Rule)")

plt.xlabel("Customers (Ranked by Revenue)")

plt.ylabel("Cumulative Revenue %")

plt.grid(True)

plt.show()
```

### Interpretation

The Pareto chart shows the cumulative percentage of revenue contributed by customers ranked from highest to lowest spending.

It helps identify whether a relatively small number of customers account for the majority of the company's revenue.

### Business Insight

If the analysis shows that approximately 20% of customers contribute around 80% of total revenue, the company should prioritize retaining these high-value customers through loyalty programs, personalized marketing, and premium customer service. Protecting these relationships can have a significant impact on overall business performance.

# RFM Customer Segmentation

```python
# Find the latest invoice date
snapshot_date = df['InvoiceDate'].max()

print(snapshot_date)
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalSales': 'sum'
})

rfm.columns = ['Recency', 'Frequency', 'Monetary']

rfm.head()
```

```python
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

rfm.head()
```

```python
rfm['RFM_Score'] = (
    rfm['R_Score'].astype(str) +
    rfm['F_Score'].astype(str) +
    rfm['M_Score'].astype(str)
)

rfm.head()
```

### Interpretation

The RFM model evaluates customers based on how recently they purchased, how frequently they purchase, and how much they spend. Higher scores indicate more valuable and engaged customers.

### Business Insight

RFM analysis enables the business to identify high-value customers, improve customer retention strategies, and design personalized marketing campaigns. Instead of treating all customers equally, the company can tailor its approach to different customer segments.
